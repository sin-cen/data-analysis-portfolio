```python
# 기본
import pandas as pd
import numpy as np

# StringIO (CSV 문자열을 파일처럼 읽기)
from io import StringIO

# 시각화
import matplotlib.pyplot as plt
import seaborn as sns

# (선택) 한글 폰트/마이너스 깨짐 방지 (윈도우 기준)
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

# (선택) seaborn 스타일
sns.set_style('whitegrid')
```

```python
df='''order_date,user_id,device,gender,age,price,qty,purchased,stay_time
2025-01-01,U001,mobile,M,23,12000,1,1,180
2025-01-01,U002,pc,F,35,54000,1,1,420
2025-01-01,U003,mobile,F,29,0,0,0,60
2025-01-02,U004,pc,M,41,67000,1,1,510
2025-01-02,U005,mobile,F,33,21000,1,1,240
2025-01-02,U006,pc,M,27,0,0,0,90
2025-01-03,U007,mobile,M,52,0,0,0,75
2025-01-03,U008,pc,F,38,83000,1,1,560
2025-01-03,U009,mobile,F,45,15000,1,1,260
2025-01-04,U010,pc,M,31,0,0,0,110
2025-01-04,U011,mobile,M,28,29000,1,1,300
2025-01-04,U012,mobile,F,34,0,0,0,80
'''
```

```python
df1=pd.read_csv(StringIO(df))
```

```python
df1
```

```python
df1['order_date']=pd.to_datetime(df1['order_date'])
```

```python
df1.info()
```

```python
df1.head(3)
```

```python
df1['총매출']=df1['price']*df1['qty']
```

```python
df1
```

```python
df1.info()
```

```python
df1.loc[(df1['age']>=20) & (df1['age']<30),'연령대']='20대'

df1.loc[(df1['age']>=30) & (df1['age']<40),'연령대']='30대'

df1.loc[(df1['age']>=40) & (df1['age']<50),'연령대']='40대'

```

```python
df1.loc[df1['age']>=50,'연령대']='50대'
```

```python
df1
```

```python
df2 = (
    df1
    .groupby('device')
    .agg(
        방문자수=('user_id', 'nunique'),
        구매자수=('purchased', 'sum'),
        구매전환율=('purchased', 'mean'),
        평균주문금액=('price', lambda x: x[x > 0].mean())
        ,평균체류시간=('stay_time','mean'))
    .reset_index()
)
```

```python
df2
```

```python
result=df1.groupby(['device','연령대']).agg(구매전환률=('purchased','mean'),
                                 총매출=('총매출','sum'),
                                 평균체류시간=('stay_time','mean')).reset_index()
```

```python
result['구매전환률']=result['구매전환률'].sort_values(ascending=False)
```

```python
df2['평균체류시간']=df2['평균체류시간'].round(2)
```

```python
df1['객단가']=df1['총매출']/12
```

```python
sns.set_style('whitegrid')              # 그리드 있는 깔끔한 스타일
plt.rc('font', family='Malgun Gothic')  # 한글 폰트(윈도우: 맑은 고딕)
plt.rc('axes', unicode_minus=False)  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(
    nrows=1,
    ncols=3,
    figsize=(18, 5)
)
ax1 = axes[0]
ax2 = axes[1]
ax3 = axes[2]

sns.barplot(data=df2,x='device',y='구매전환율',ax=ax1)
ax1.set_title('기기별 구매 전환율')
sns.barplot(data=df2,x='device',y='평균주문금액',ax=ax2)
ax2.set_title('기기별 평균주문금약')
sns.scatterplot(data=df1,
               x='stay_time',
               y='price',
               hue='purchased',
               ax=ax3)
ax3.set_title('구매자 비구매자별 페이지시간 ')
```
