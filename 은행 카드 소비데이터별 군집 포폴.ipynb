```python
ratio_df=pd.concat([ratio,ratio1],axis=1)
```

```python
ratio_df.columns=['2022','2025']
```

```python
ratio_df.plot(
kind='bar',
figsize=(10,6))

plt.title('업종별 소비 비중 비교(2022 vs 2025)')
plt.ylabel('소비 비중')
plt.xticks(rotation=45)
plt.legend(title='연도')
plt.tight_layout()
plt.show()
```

```python
ratio_df['변화량(2025-2022)'] = ratio_df['2025'] - ratio_df['2022']

ratio_df.sort_values('변화량(2025-2022)', ascending=False)
```

```python
ratio_df
```

```python
cluster_df=ratio_df[['2022','2025','변화량(2025-2022)']]
cluster_df
```

```python
from sklearn.preprocessing import StandardScaler   

scaler = StandardScaler()
X_scaled=scaler.fit_transform(cluster_df)
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3,
             random_state=42)

cluster_labels=kmeans.fit_predict(X_scaled)

cluster_df['cluster']=cluster_labels
cluster_df
```

```python
cluster_summary = (
    cluster_df
    .groupby('cluster')
    .mean()
)

cluster_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

for c in cluster_df['cluster'].unique():
    subset = cluster_df[cluster_df['cluster'] == c]
    plt.scatter(
        subset['2025'],
        subset['변화량(2025-2022)'],
        label=f'Cluster {c}',
        s=100
    )

plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('2025 소비 비중')
plt.ylabel('소비 비중 변화량')
plt.title('업종별 소비 구조 변화 군집')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
