```python
#지표	계산 방법
#방문자 수	visited 합
#구독자 수	subscribed 합
#전환율	구독자 수 / 방문자 수
#총매출	매출 합
#ARPU	총매출 / 방문자 수

import streamlit as st
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

# ── 한글 폰트 설정 ──
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
mpl.rcParams["font.family"] = font_name
mpl.rcParams["axes.unicode_minus"] = False
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO

df='''date,plan,user_id,is_new,visited,subscribed,price
2025-03-01,basic,U001,1,1,0,10000
2025-03-01,premium,U002,1,1,1,30000
2025-03-01,basic,U003,0,1,1,10000
2025-03-02,premium,U004,1,1,1,30000
2025-03-02,basic,U005,1,1,0,10000
2025-03-02,premium,U006,0,1,0,30000
2025-03-03,basic,U007,0,1,1,10000
2025-03-03,premium,U008,1,1,1,30000
2025-03-04,basic,U009,1,1,0,10000
2025-03-04,premium,U010,0,1,1,30000'''

df1=pd.read_csv(StringIO(df))

df1['date']=pd.to_datetime(df1['date'])

df1['요일']=df1['date'].dt.day_name()

df1['매출']=df1['subscribed']*df1['price']

df2=df1.groupby(['plan','is_new']).agg(방문자수=('visited','sum'),
                                  구독자수=('subscribed','sum'),
                                  총매출=('매출','sum'))

df2=df2.reset_index()

df2['전환율']=df2['구독자수']/df2['방문자수']

df2['ARPU']=df2['총매출']/df2['방문자수']

df1['요일']=df1['date'].dt.day_name()

df3=df1.groupby(['요일','plan']).agg(총매출=('매출','sum'),
                                  방문자수=('visited','sum'),
                                  구독자수=('subscribed','sum'),
                                  신규유입=('is_new','sum'))

df3['전환율']=df3['구독자수']/df3['방문자수']

df3=df3.reset_index()

#######타이틀


st.title('유튜브 구독자 분석 정보')



st.subheader('요일별 분석정보')
st.dataframe(df3)

fig1, ax1 = plt.subplots()
ax1.bar(df3['요일'],df3['전환율'])
ax1.set_title('요일별 전환율')
st.pyplot(fig1)

fig2, ax2 = plt.subplots()
ax2.bar(df3['요일'],df3['총매출'])
ax2.set_title('요일별 총매출')
st.pyplot(fig2)


st.subheader('plan별 신규유입 구독자 방문자 ARPY 분석정보')
st.dataframe(df2)

fig3, ax3 = plt.subplots()
ax3.bar(df2['plan'],df2['방문자수'])
ax3.set_title('플랜별 방문자수')
st.pyplot(fig3)

fig4, ax4 = plt.subplots()
ax4.bar(df2['plan'],df2['ARPU'])
ax4.set_title('플랜별 ARPU')
st.pyplot(fig4)

st.metric("최고 ARPU ",'premium')
st.metric("최고 전환율 플랜 ",'premium')
```
